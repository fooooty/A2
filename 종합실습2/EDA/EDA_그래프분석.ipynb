```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rc('font',family='NanumGothic ECO')
plt.rc('axes',unicode_minus=False)
```

```python
df = pd.read_csv("/home/piai/second_week/2. Big Data 분석/preprocessed_scale.csv")
df
```

```python
plt.figure(figsize=(6, 4))
p1 = sns.countplot(x='steel_kind', hue='scale', data=df)

# 각 막대 위에 카운트 수 표시
for p in p1.patches:
    p1.annotate(format(int(p.get_height()), 'd'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5), textcoords='offset points')

plt.title('Count of Scale by Steel Kind')
plt.xlabel('Steel Kind')
plt.ylabel('Count')
plt.legend(title='Scale')
plt.show()
```
